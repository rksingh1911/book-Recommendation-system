```python
 import ast
# for i in ast.literal_eval(obj):
```

```python
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['cast'][0]
```

```python
def convert2(obj):
    l = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            l.append(i['name'])
            count += 1
        else:
            break
        
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
        
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words='english')
```

```python
vector =  cv.fit_transform(new_df['tags']).toarray()
```

```python
#cv.get_feature_names()
```

```python
#cosine similarity is the angle between targat words from rest for the words here we can't calculate eucliden distance because this calculation will be very long
#and also req very high camputanal power
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x : x[1]) #[1:6]
#here we are cheking from one single movie what is the similarity score to other 4805 movie and sorting it 
```

```python
similarity[1]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distnace = similarity[movie_index]
    movies_list = sorted(list(enumerate(distnace)),reverse=True,key=lambda x : x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```
